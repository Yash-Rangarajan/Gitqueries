```python
#public repository retrieval
for repository in github3.all_repositories(number=10):
    print('{0}, id: {0.id}, url: {0.html_url}'.format(repository))
```

```python
#repositories created by a specific user
for repository in github3.repositories_by('sigmavirus24'):
    print('{0} created at {0.created_at}'.format(repository))
```

```python
#retrive issues in repository

repo = github3.repository('sigmavirus24', 'github3.py')

for issue in repo.issues():
    print('{0}#{1.number}: "{1.title}"\n\t{1.html_url}'.format(repo, issue))
    
```

```python
print("CLOSED ISSUES")
for issue in repo.issues(state='closed'):
    print('{0}#{1.number}: "{1.title}"\n\t{1.html_url}'.format(repo, issue))
```

```python

#repo interactions: Forks, starrers, watchers, open issues

print(repo.forks_count)
print(repo.stargazers_count)
print(repo.subscribers_count)
print(repo.open_issues_count)
```

```python
#issues,comments and labels
issue = github3.issue('sigmavirus24', 'github3.py', 187)

print(issue.body+ "\n")

print(issue.comments_count)
print("\n")
for comment in issue.comments():
    print('"{0.body_text}" posted by {0.user}'.format(comment))
    print('-' * 80)
print("LABELS")
for label in issue.labels():
    print(label)
```
